```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
# Iris
iris = load_iris()
X = iris.data[:, [0, 1, 2, 3]]  
```

```python
# modèle K-means avec 2 clusters
kmeans_2 = KMeans(n_clusters=2)
kmeans_2.fit(X)
y_pred_2 = kmeans_2.predict(X)

```

```python
# Entraînement du modèle K-means avec 3 clusters
kmeans_3 = KMeans(n_clusters=3)
kmeans_3.fit(X)
y_pred_3 = kmeans_3.predict(X)
```

```python
# Entraînement du modèle K-means avec 4 clusters
kmeans_4 = KMeans(n_clusters=4)
kmeans_4.fit(X)
y_pred_4 = kmeans_4.predict(X)
```

```python

# Affichage des graphiques pour 2 clusters
plt.figure(figsize=(12, 4))

plt.subplot(121)
plt.scatter(X[:, 0], X[:, 1], c=y_pred_2)
plt.scatter(kmeans_2.cluster_centers_[:, 0], kmeans_2.cluster_centers_[:, 1], marker='x', color='red')
plt.xlabel('Longueur de pétale')
plt.ylabel('Largeur de pétale')
plt.title('Clustering des iris avec 2 clusters')

# Affichage des graphiques pour 3 clusters
plt.subplot(122)
plt.scatter(X[:, 0], X[:, 1], c=y_pred_3)
plt.scatter(kmeans_3.cluster_centers_[:, 0], kmeans_3.cluster_centers_[:, 1], marker='x', color='red')
plt.xlabel('Longueur de pétale')
plt.ylabel('Largeur de pétale')
plt.title('Clustering des iris avec 3 clusters')

# Affichage des graphiques pour 4 clusters
#plt.subplot(122)
#plt.scatter(X[:, 0], X[:, 1], c=y_pred_4)
#plt.scatter(kmeans_4.cluster_centers_[:, 0], kmeans_4.cluster_centers_[:, 1], marker='x', color='red')
#plt.xlabel('Longueur de pétale')
#plt.ylabel('Largeur de pétale')
#plt.title('Clustering des iris avec 4 clusters')

plt.tight_layout()
plt.show()
```

```python
# Évaluation du modèle avec la méthode Elbow
inertia = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(range(2, 11), inertia, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Inertie')
plt.title('Méthode Elbow')
plt.show();

```

```python
# Évaluation du modèle avec la méthode Silhouette

silhouette_scores = []
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    y_pred = kmeans.predict(X)
    silhouette_scores.append(silhouette_score(X, y_pred))

plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.xlabel('Nombre de clusters')
plt.ylabel('Score de silhouette')
plt.title('Méthode Silhouette')
plt.show()
```

```python
# Nombre de clusters optimal selon la méthode Elbow
optimal_clusters_elbow = np.argmin(np.diff(inertia)) + 2
print(f"Nombre de clusters optimal selon la méthode Elbow : {optimal_clusters_elbow}")

# Nombre de clusters optimal selon la méthode Silhouette
optimal_clusters_silhouette = np.argmax(silhouette_scores) + 2
print(f"Nombre de clusters optimal selon la méthode Silhouette : {optimal_clusters_silhouette}")
```
