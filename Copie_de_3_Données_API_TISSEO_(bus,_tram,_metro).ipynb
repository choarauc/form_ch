Vous pouvez utiliser la bibliothèque requests en Python pour effectuer des requêtes GET à l'API. Voici un exemple simple pour récupérer les horaires à un arrêt spécifique en temps réel :

Paramètres possibles :
- **STOP_AREAS : ZONES D’ARRETS**
- **STOP_POINTS : ARRETS (POTEAUX D’ARRETS)**
- PLACES : RECHERCHE DE LIEUX ET GEOCODAGE
- NETWORKS : RESEAUX DE TRANSPORT DISPONIBLES
- **LINES : LES LIGNES COMMERCIALES**
- **STOPS_SCHEDULES : PROCHAINS PASSAGES**
- **ROLLING_STOCKS : MODES DE TRANSPORTS**
- JOURNEYS : CALCUL D’ITINERAIRES
- **MESSAGES : MESSAGES D’INFORMATION**

- Les Lines sont des lignes commerciales (exemple : ligne T1, ligne 2, …).
- Une Line appartient à un réseau de transport : network. Dans l’API le network vaudra toujours « Tisséo ».
Cependant, l’objet « network » a été ajouté en prévision d’une ouverture à d’autres réseaux.
- Une Line a un transportMode ou RollingStock : bus, métro, tramway, transport à la demande.
- Chaque Line est composée d’un ou de plusieurs itinéraires, généralement au moins un aller retour, mais
parfois plus pour les lignes à fourches ou à parcours spécifiques.
- Un itinéraire dessert des StopPoints (ou PhysicalStops) qui correspondent à des poteaux d’arrêts.
- Dans le cas général, il y a au moins deux StopPoints face à face (un de chaque côté de la rue pour
chaque sens) regroupés dans un Stop Area (ou Arrêt logique).
Un Stop Area porte la dénomination commerciale. C’est à cet objet « nommable » (voir 3.12) qu’un
usager fait allusion lorsqu’il dit « Je pars d’Esquirol ».
- Des messages d’information trafic peuvent affecter des Lines ou être globales (Voir 4.10)

```python
import requests
import pandas as pd

service_name = 'stop_points'
format_type = 'json'
parameters = ''
api_key = '15cbfcdf-76bb-4136-980a-6dc1f1d96cd5'


# modele d'url = https://api.tisseo.fr/v2/<nom du service>.<format>?<paramètres>&key=<'15cbfcdf-76bb-4136-980a-6dc1f1d96cd5'>
url = f"https://api.tisseo.fr/v2/{service_name}.json?&key={api_key}"

# url = "https://api.tisseo.fr/v2/stop_points.json?&key=15cbfcdf-76bb-4136-980a-6dc1f1d96cd5"



response = requests.get(url)

if response.status_code == 200:
    data = response.json()

    # Récupération des enregistrements
    records = data['physicalStops']

    df_stop_points = []
    for record in records['physicalStop']:
        id_stop_point = record["id"]
        stop_point_name = record["name"]
        #short_name_ligne = record['lines']['short_name']
        id_stop_area = record['stopArea']['id']

        df_stop_points.append({
            "id_stop_point": id_stop_point,
            "stop_point_name": stop_point_name,
            #"short_name_ligne": short_name_ligne,
            "id_stop_area": id_stop_area
        })

    df_stop_points_final = pd.DataFrame(df_stop_points)
else:
    print("Erreur lors de la requête à l'API")
```
