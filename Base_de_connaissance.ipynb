## 5. Les Tuples

 Les tuples (t-uplets en franais) sont une structure de donnes qui ressemble aux listes :
>
> *#Cration d'un tuple*
>
>`un_tuple = ("Bonjour", -1, 133)`
>
>`un_tuple = "Bonjour", -1, 133`
>
> *# Affichage du premier lment du tuple*
>
> `print(un_tuple[0])`
>
> *# Affichage du dernier lment du tuple*
>
> `print(un_tuple[-1])`
>
>> * La dfinition d'un tuple se fait **avec ou sans parenthses.**
>>
>> * L'indexation d'un tuple est **identique**  celle d'une liste.
>
> Une nuance trs importante des tuples est **qu'ils ne sont pas modifiables.**
>
> La puissance des tuples ne sera pas vidente tout de suite, mais l'un des atouts majeurs de cette classe est ce qu'on appelle le **tuple assignement** (assignation par un tuple en franais), qui permet **d'assigner des valeurs  plusieurs variables simultanment :**

````python
un_tuple = "Bonjour", -1, 133
print(un_tuple)

# Tuple assignment
x, y, z = un_tuple

print(x)
print(y)
print(z)
````

>Les variables x, y et z ont t cres et simultanment assignes  des valeurs. Pour que le tuple assignment se droule correctement, il faut qu'il y ait autant de variables  assigner que d'lments dans le tuple.
>
>Le tuple assignment donne une solution syntaxique lgante au problme d'change de valeurs : Nous avons deux variables a et b et nous voulons changer leurs valeurs, c'est--dire que a doit prendre la valeur de b et b la valeur de a.
>
>Dans des langages de programmation plus classiques, nous serions obligs de crer une variable temporaire qui contient une des valeurs de a ou b:
>
>* *# On stocke la valeur de a dans une variable temporaire*
>
>tmp = a
>
>* *# On met  jour a avec la valeur de b*
>
>a = b
>
>* *# On met  jour b avec la valeur de la variable temporaire*
>
>b = tmp
>
>Grce au tuple assignement, cette opration peut se faire en une ligne :
>
>* *# change de valeurs entre a et b*
>
>a, b = b, a

````python
a = 10
b = 5

a, b = b, a

print(a)
print(b)
````

````python
# calculer_total, prend en entre le prix et la quantit du produit, calcule le total de la commande en multipliant le prix par la quantit, et applique une rduction de 10% si le total est suprieur  100 euros. La fonction retourne le total de la commande.

def calculer_total(prix, quantite):
    total = prix * quantite
    if total > 100:
        total *= 0.9
    return total

# ajouter_frais_livraison, prend en entre le total de la commande et le pays de destination, et ajoute les frais de livraison correspondants en fonction du pays de destination. La fonction retourne le total de la commande avec les frais de livraison ajouts.

def ajouter_frais_livraison(total, pays):
    if pays == "France":
        frais_livraison = 5
    elif pays == "Espagne":
        frais_livraison = 10
    elif pays == "Etats-Unis":
        frais_livraison = 15
    elif pays == "Australie":
        frais_livraison = 20
    else:
        frais_livraison = 0
    return total + frais_livraison

# passer_commande, prend en entre le prix du produit, la quantit commande et le pays de destination, et calcule le total de la commande en utilisant les deux fonctions prcdentes. La fonction retourne le total de la commande.

def passer_commande(prix, quantite, pays):
    total = calculer_total(prix, quantite)
    total = ajouter_frais_livraison(total, pays)
    return total


# test < 100

prix_produit = 25
quantite = 3
pays_destination = "France"

total_commande = passer_commande(prix_produit, quantite, pays_destination)
print("Le total de la commande est de", total_commande, "euros.")


# test > 100

prix_produit = 150
quantite = 1
pays_destination = "France"

total_commande = passer_commande(prix_produit, quantite, pays_destination)
print("Le total de la commande est de", total_commande, "euros.")
````

````python
#@title
roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
````

````python
#@title
def roman_to_arabic(roman_numeral):
    '''
    La fonction prend en entre une chane de caractres roman_numeral reprsentant un chiffre romain.

    La fonction cre un dictionnaire roman_dict qui associe chaque chiffre romain  sa valeur en chiffre arabe.

    La variable arabic_num est initialise  0, elle va contenir la valeur en chiffre arabe correspondant au chiffre romain.

    La variable prev_value est initialise  0, elle va contenir la valeur du chiffre romain prcdent.

    La boucle for parcourt les caractres de la chane roman_numeral de droite  gauche en utilisant la mthode [::-1] pour inverser l'ordre de la chane.

    Pour chaque caractre char dans la chane roman_numeral, on rcupre sa valeur en chiffre arabe depuis le dictionnaire roman_dict.

    Si la valeur current_value est infrieure  la valeur prev_value, cela signifie qu'il faut soustraire current_value  arabic_num. Sinon, on ajoute simplement current_value  arabic_num.

    On met ensuite  jour la valeur de prev_value avec current_value.
    '''
    roman_dict = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}
    arabic_num = 0
    prev_value = 0 
    for char in roman_numeral[::-1]:
        current_value = roman_dict[char]        
        if current_value < prev_value:
            arabic_num -= current_value
        else:
            arabic_num += current_value
        
        prev_value = current_value
        
    return arabic_num
````

````python
#@title
import numpy as np

class Complexe:
    def __init__(self, a = 0, b = 0):
        self.partie_re = a
        self.partie_im = b
    
    def __str__(self):
        if(self.partie_im < 0):
            return self.partie_re.__str__() + self.partie_im.__str__() + 'i'
        
        if(self.partie_im == 0):
            return self.partie_re.__str__() 
        
        if(self.partie_im > 0):
            return self.partie_re.__str__() + '+' + self.partie_im.__str__() + 'i' 
        
    def mod(self):
        return np.sqrt( self.partie_re ** 2 + self.partie_im ** 2)  # renvoie (sqrt(a + b))
        
# Le paramtre 'other' correspond  l'objet de type Complexe avec lequel nous voulons effectuer
# la comparaison
    
    def __lt__(self, other):    
        if( self.mod() < other.mod()):   # renvoie True si |self| < |other|
            return True
        else:
            return False
        
    def __gt__(self, other):
        if( self.mod() > other.mod()):   # renvoie True si |self| > |other|
            return True
        else:
            return False
    
z1 = Complexe(3, 4)
z2 = Complexe(2, 5)

print(z1 < z2)
print(z1 > z2)

print(z1)
````

````python
#@title

def puissanceA(n):
    A = np.array([[1, 0],
                [0, 1]])

    B = np.array([[1, -1],
                  [-1, 1]])
    
    for i in range(n):
        A = A.dot(B)
        
    return A

print("A**2: \n", puissanceA(2), "\n")
print("A**3: \n", puissanceA(3), "\n")
print("A**4: \n", puissanceA(4), "\n")
````

### Broadcasting entre une matrice et un vecteur

* (a) Dfinir une fonction nomme normalisation_min_max prenant en argument une matrice  X
  et qui retournera  X 


+ (b) Appliquer la normalisation Min-Max sur  X

 . Vous devriez obtenir  deux dcimales prs la matrice suivante 



```
    1    1
x = 0.4 0.13    
    0    0
```



````python
#@title
def normalisation_min_max(X):
      
    X_tilde = np.zeros(shape = X.shape)
    
    # Pour chaque colonne de X
    for j, column in enumerate(X.T):
        # Initialisation du minimum et du maximum de la colonne
        min_Xj = column[0]
        max_Xj = column[0]
        
        # Pour chaque valeur dans la colonne
        for value in column:
            # Si la valeur est plus PETITE que le min
            if value < min_Xj:
                # On crase le min avec cette valeur
                min_Xj = value
            
            # Si la valeur est plus GRANDE que le max
            if value > max_Xj:
                # On crase le max avec cette valeur
                max_Xj = value
                
        # On peut maintenant calculer X_tilde pour cette colonne
        # Le broadcasting nous permet de faire cette opration sans boucle for
        X_tilde[:, j] = (X[:, j] - min_Xj) / (max_Xj - min_Xj)
            
    return X_tilde

X = np.array([[24, 1.88],
              [18, 1.68],
              [14, 1.65]])

X_tilde = normalisation_min_max(X)

print(X_tilde)
````

La formule de l'erreur quadratique moyenne, s'abrge par  MSE

Dans le cas de la rgression linaire, l'objectif de l'erreur quadratique moyenne est de trouver le vecteur de rgression

  * (a) Dfinir une fonction nomme mean_squared_error prenant en argument une matrice X, un vecteur beta et un vecteur y et qui, sans boucle for, retourne l'erreur quadratique moyenne associe.

````python
#@title

# (a) Dfinir une fonction nomme mean_squared_error prenant en argument une matrice X, un vecteur beta et un vecteur y et qui, sans boucle for, retourne l'erreur quadratique moyenne associe.

def mean_squared_error(X, beta, y):
    # Calcul de ^y
    y_hat = X.dot(beta)
    
    # Calcul de l'erreur quadratique (^y_i - y_i)**2
    mse = (y_hat - y)**2
    
    # On fait la moyenne de l'erreur pour obtenir la MSE
    mse = mse.mean()
    
    return mse
````

>Une transaction ACID est une transaction qui respecte les points suivants :
>>
>>+ Atomicit : une transaction doit tre effectue en entier ou abandonne, c'est  dire que si une opration est refuse, toutes les oprations de la transaction ne seront pas valides.
>>
>> + Cohrente : l'tat initial et final doivent tre cohrents, c'est  dire respecter les contraintes d'intgrit.

La syntaxe pour ajouter un tuple dans une table est la suivante :

`INSERT INTO nom_table VALUES tuple`

Cette commande est en langage SQL, et donc sera au format texte sous python si on souhaite l'excuter. La valeur doit donc tre spcifie via un paramtre associ. On note 'INSERT OR REPLACE INTO students VALUES ({markers})' o markers sera une variable explicitant le format des tuples que l'on injectera dans la table.


Une fois le markers dfinit, on peut mettre  jour notre requte SQL en texte en modifiant son format, c'est pour que le markers ne soit pas considr comme une chane de caractres mais bien un "argument". ````python
# on cre la connection
with engine.connect() as connection:
    # dbut de la transaction
    with connection.begin() as transaction:
        # on tente d'xcuter une transaction
        try:
            # On indique le format d'un tuple de cette table
            markers = ','.join('?' * len(values[0])) 
            
            # On utilise le langage SQL en format texte o markers est le format d'un tuple
            ins = 'INSERT OR REPLACE INTO {tablename} VALUES ({markers})'
            
            # On prcise ce format particulier grce  la fonction membre format
            ins = ins.format(tablename=students.name, markers=markers)
           
            # Enfin on peut utiliser les tuples crs en xcutant la commande SQL
            connection.execute(ins, values)
        # si la transaction choue
        except:
            transaction.rollback()
            raise
        # si la transaction russit
        else:
            transaction.commit()
````
